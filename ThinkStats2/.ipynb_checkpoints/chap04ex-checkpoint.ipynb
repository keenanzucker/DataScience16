Read the pregnancy file.

```python
%matplotlib inline

import nsfg
preg = nsfg.ReadFemPreg()
```

Select live births, then make a CDF of <tt>totalwgt_lb</tt>. 

```python
import thinkstats2

live = preg[preg.outcome == 1]
firsts = live[live.birthord == 1]
others = live[live.birthord != 1]

cdf = thinkstats2.Cdf(live.totalwgt_lb)
```

Display the CDF.

```python
import thinkplot

thinkplot.Cdf(cdf, label="CDF of birthweight")
thinkplot.Show(xlabel="weight in pounds", ylabel="CDF", title="CDF vs Weight")
```

Find out how much you weighed at birth, if you can, and compute CDF(x). 

```python
my_weight = 9.1

print "My weight Probability: " , cdf.Prob(my_weight)

thinkplot.Cdf(cdf)
thinkplot.Plot(my_weight, cdf.Prob(my_weight), 'o', markersize=20, label="Keenan")
thinkplot.Show(xlabel="weight in pounds", ylabel="CDF", title="CDF vs Weight plus Keenan")
```

If you are a first child, look up your birthweight in the CDF of first children; otherwise use the CDF of other children.

```python
cdf = thinkstats2.Cdf(others.totalwgt_lb)
print 'Probability of my weight as non-first' , cdf.Prob(my_weight)

thinkplot.Cdf(cdf)
thinkplot.Plot(my_weight, cdf.Prob(my_weight), 'o', markersize=20, label="Keenan")
thinkplot.Show(xlabel="weight in pounds", ylabel="CDF", title="CDF vs Weight plus Keenan")
```

Compute the percentile rank of your birthweight

```python
print "My percentage Rank: ", cdf.PercentileRank(my_weight), "%"
```

Compute the median birth weight by looking up the value associated with p=0.5.

```python
cdf = thinkstats2.Cdf(live.totalwgt_lb)

print "median birth weight: ", cdf.Value(0.5)
```

Compute the interquartile range (IQR) by computing percentiles corresponding to 25 and 75. 

```python
p25 = cdf.Percentile(25)
p75 = cdf.Percentile(75)

print "Interquartile Range is between %s and %s pounds" % (p25, p75)
```

Make a random selection from <tt>cdf</tt>.

```python
print "Random Selection: %s" % cdf.Random()
```

Draw a random sample from <tt>cdf</tt>.

```python
sample = cdf.Sample(100)

print sample
```

Draw a random sample from <tt>cdf</tt>, then compute the percentile rank for each value, and plot the distribution of the percentile ranks.

```python
ranks = [cdf.PercentileRank(x) for x in sample]
rank_cdf = thinkstats2.Cdf(ranks)
thinkplot.Cdf(rank_cdf)
thinkplot.Show(xlabel='percentile rank', ylabel='CDF')
```

Generate 1000 random values using <tt>random.random()</tt> and plot their PMF.

```python
import numpy.random as np

values = np.random(1000)

pmf = thinkstats2.Pmf(values, label="PMF of random values")
thinkplot.Pmf(pmf)
thinkplot.Show()
```

Assuming that the PMF doesn't work very well, try plotting the CDF instead.

```python
cdf = thinkstats2.Cdf(values, label="CDF of random values")
thinkplot.Cdf(cdf)
thinkplot.Show()
```
