Suppose one of your co-workers is expecting a baby and you are participating in an office pool to predict the date of birth. Assuming that bets are placed during the 30th week of pregnancy, what variables could you use to make the best prediction? You should limit yourself to variables that are known before the birth, and likely to be available to the people in the pool. 

______
I need to first gather data, as well as only include the columns that happen at the 30 week pregnancy length and longer

```python
import numpy as np
import thinkstats2
import thinkplot
import nsfg
import statsmodels.formula.api as smf

df = nsfg.ReadFemPreg()
df = df[df.prglngth>=30]
df
```

I'm going to split up the data to only include the variables that I think could possibly make an impact on the pregnancy length, which I decided are age, race, birth order, and weight. I made a new dataframe that only contains this data as well as the actual pregnancy lenght to train. I also dropped the NA values.

```python
df = df.dropna(subset=['prglngth', 'agepreg', 'race', 'birthord', 'basewgt'])
newdf = df[['prglngth', 'agepreg', 'race', 'birthord', 'basewgt']].copy()

lengths = df.prglngth
ages = df.agepreg
races = df.race
orders = df.birthord
weights = df.basewgt

newdf.groupby('prglngth').mean()
```

I started by comparing the dependent variable, pregnancy length, with age pregnant

```python
formula1 = 'prglngth ~ agepreg'
model = smf.ols(formula1, data=newdf)
results = model.fit()
print results.summary()
```

I then compared the dependent variable, pregnancy length, with race

```python
formula1 = 'prglngth ~ race'
model = smf.ols(formula1, data=newdf)
results = model.fit()
print results.summary()
```

I then compared the dependent variable, pregnancy length, with birth order of the child

```python
formula1 = 'prglngth ~ birthord'
model = smf.ols(formula1, data=newdf)
results = model.fit()
print results.summary()
```

I then compared the dependent variable, pregnancy length, with respondents weight

```python
formula1 = 'prglngth ~ basewgt'
model = smf.ols(formula1, data=newdf)
results = model.fit()
print results.summary()
```
